```python
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

from torchvision import models
from torchvision import transforms

import matplotlib.pyplot as plt
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(device)
```

### Функции для обработки фото

```python
# pipeline для обработки фото: сжатие и приведение к тензорному формату
loader = transforms.Compose([
                            transforms.Resize((512, 512)),
                            transforms.ToTensor()
                            ])

def image_loader(image_path):
    # открываем файл с диска
    image = Image.open(image_path)
    # добавляем нулевую батч-размерность
    image = loader(image).unsqueeze(0)
    # отправляем в cuda, если есть
    return image.to(device)
    # return image.to(device, torch.float)

def imshow(tensor, title=None):
    # возвращаем на cpu и убираем первую размерность
    image = tensor.clone().cpu()
    image = image.squeeze(0)
    # возвращаем формат фотки
    image = transforms.ToPILImage()(image)
    plt.imshow(image)
    if title is not None:
        plt.title(title)
```

```python
path = '/content/gdrive/MyDrive/NST/'
style1 = image_loader(path + "style1.jpg")
style2 = image_loader(path + "style2.jpg")
content = image_loader(path + "content.jpg")
```

```python
fig = plt.figure(figsize=(17, 10))
plt.subplot(131)
imshow(style1, title='The Shipwreck of the Minotaur by J.M.W. Turner, 1805')
plt.subplot(133)
imshow(style2, title='Femme nue assise by Pablo Picasso, 1910')
plt.subplot(132)
imshow(content, title='Depp and Heard')
```

### Функции для создания лосса и создание класса для получения фичей

```python
def content_loss(generated, content):
    return torch.mean((generated - content)**2)

def style_loss(generated, style):
    batch_size, channel, height, width = generated.shape

    # изменим shape матриц на формат "количество каналов x высота * ширина"
    generated_reshaped = generated.view(channel, height*width)
    style_reshaped = style.view(channel, height*width)

    # посчитаем матрицы Грамма
    G = torch.mm(generated_reshaped, generated_reshaped.t())
    S = torch.mm(style_reshaped, style_reshaped.t())
    
    # и итоговый лосс
    return torch.mean((S-G)**2)

def calculate_loss(generated_features, content_features, style_features, alpha=10, beta=50):
    cont_loss, st_loss = 0, 0

    for gen, cont in zip(generated_features, content_features):
        cont_loss += 0.2 * content_loss(gen, cont)
    
    for gen, st in zip(generated_features, style_features):
        st_loss += 0.2 * style_loss(gen, st)
    
    total_loss = alpha * cont_loss + beta * st_loss 
    return total_loss

# следующая идея пути получения фичей взята из данного репо:
# https://github.com/Octaves0911/Neural_Style_Transfer/blob/master/NST.py
# (верхняя идея с zip оттуда же, разбил на два цикла думая, что буду рассматривать разное количество слоёв 
# (для content обычно хватает фичей с 1-2 свёрточных слоёв), но потом передумал)

class VGG(nn.Module):
    def __init__(self):
        super(VGG, self).__init__()
        # Для нашей сети понадобится обученная заранее свёрточная нейросеть: классически используют VGG19
        # (переведём сразу на cuda и в боевой режим)
        self.model = models.vgg19(pretrained=True).features.to(device).eval()[:29]
        # Также выберем слои, откуда будем брать низкоуровневые фичи для нашей модели
        # (для удобства, и для переноса контента и для переноса стиля аккумулируем информацию с нескольких слоёв)
        self.required_features = [0, 7, 14, 21, 28]

    # x holds the input tensor(image) that will be feeded to each layer
    def forward(self, x):
        # initialize an array that wil hold the activations from the chosen layers
        features = []
        # Iterate over all the layers of the mode
        for layer_num,layer in enumerate(self.model):
            # activation of the layer will stored in x
            x = layer(x)
            # appending the activation of the selected layers and return the feature array
            if layer_num in self.required_features:
                features.append(x)
        
        return features
```

### Тест №1

```python
# инициализация модели
model = VGG()

# возьмём за изначальную картинку картинку с контентом
generated = content.clone().requires_grad_(True)
# Адам с темпом обучения 0.002
optimizer = optim.Adam([generated], lr=0.002)

# обучим модель
for epoch in range(10000):
    # получаем признаки
    generated_features = model(generated)
    content_features = model(content)
    style_features = model(style1)
    
    # считаем функцию потерь
    loss = calculate_loss(generated_features, content_features, style_features)
    # и оптимизируем
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # смотрим лосс на каждой 1000-ой эпохе
    if epoch % 1000 == 0:
        print(loss)
```

```python
fig = plt.figure(figsize=(17, 10))
plt.subplot(131)
imshow(style1, title='The Shipwreck of the Minotaur by J.M.W. Turner, 1805')
plt.subplot(132)
imshow(generated, title='Generated')
plt.subplot(133)
imshow(content, title='Original')
```

### Попробуем скомбинировать два стиля: подправим лосс и запустим модельку

```python
def calculate_double_loss(generated_features, content_features, style1_features, style2_features, alpha=10, beta1=40, beta2=60):
    cont_loss, st_loss1, st_loss2 = 0, 0, 0

    for gen, cont, st1, st2 in zip(generated_features, content_features, style1_features, style2_features):
        cont_loss += 0.2 * content_loss(gen, cont)
        st_loss1 += 0.2 * style_loss(gen, st1)
        st_loss2 += 0.2 * style_loss(gen, st2)
    
    total_loss = alpha * cont_loss + beta1 * st_loss1 + beta2 * st_loss2
    return total_loss
```

```python
# инициализация модели
model = VGG()

# возьмём за изначальную картинку картинку с контентом
generated = content.clone().requires_grad_(True)
# Адам с темпом обучения 0.003 (сделаем чуть более сильное обучение и снизим количество эпох далее)
# Disclaimer: Увидел слабое изменение лосса после 4000 эпох, прервал обучение и продолжил с 0.001 ещё 4000 эпох
optimizer = optim.Adam([generated], lr=0.001)

# обучим модель
for epoch in range(4000):
    # получаем признаки
    generated_features = model(generated)
    content_features = model(content)

    style1_features = model(style1)
    style2_features = model(style2)
    
    # считаем функцию потерь (больший вес по дефолту имеет Picasso: 60 против 40 у Turner - да и в целом стилю теперь осозанно придаётся больший вес)
    loss = calculate_double_loss(generated_features, content_features, style1_features, style2_features)
    # и оптимизируем
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # смотрим лосс на каждой 1000-ой эпохе
    if epoch % 1000 == 0:
        print(loss)
```
